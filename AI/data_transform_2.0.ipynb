# Reading data

```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm 
from sklearn.utils import shuffle
```

```python
PATH = os.path.join("../data", "train_scaled.csv")
data = pd.read_csv(PATH)
data = shuffle(data)
data = data.dropna()
data
```

# Changing scale

```python
## коэффициент одышки 0.4
## коэффициент остальных симптомов 0.5
# %%time
for i in (range(len(data))):
    elem = data.iloc[i]
    for j in range(len(elem[:5])):     ## changing sympthoms
        if j != 3:
            elem[j] = round(0.5 * elem[j])
        else:
            elem[j] = round(0.4 * elem[j])
    
    data.iloc[i] = elem
```

```python
data.to_csv("train_scaled_2.0.csv", index = False)
```
